#convert text to machine format


```python
from sklearn.feature_extraction.text import TfidfVectorizer

def get_skills_text(row):
    skills = []
    if row['Skills_Python'] == 1: skills.append('python')
    if row['Skills_SQL'] == 1: skills.append('sql')
    if row['Skills_ML'] == 1: skills.append('ml')
    if row['Skills_Deep_Learning'] == 1: skills.append('deep_learning')
    if row['Skills_Cloud'] == 1: skills.append('cloud')
    return ' '.join(skills) if skills else ''

df['Skills_Text'] = df.apply(get_skills_text, axis=1)

vectorizer = TfidfVectorizer()
skill_matrix = vectorizer.fit_transform(df['Skills_Text'])
```

#training and testing part

##select relevant columns


```python
df = df[['Job_Title', 'Skills_Text']]
```

##define output and input


```python
X = df['Skills_Text']   # Input (skills)
y = df['Job_Title']      # Output (job titles)
```

##import splitting tools


```python
from sklearn.model_selection import train_test_split
```

##split the data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% testing
    random_state=42     # ensures consistency
)
```

##check the split

```python
print("Training data:", len(X_train))
print("Testing data:", len(X_test))
```

##convert text to numbers

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_train_vector = vectorizer.fit_transform(X_train)
X_test_vector = vectorizer.transform(X_test)
```

##apply skill matching model

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(X_test_vector, X_train_vector)
```

##predict jobs

```python
import numpy as np

predicted_indices = similarity.argmax(axis=1)
predicted_jobs = y_train.iloc[predicted_indices].values
```

##evaluate accuracy


```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predicted_jobs)
print("Model Accuracy:", accuracy)
```

#analyse model performance

```python
comparison = pd.DataFrame({
    "Actual Job": y_test,
    "Predicted Job": predicted_jobs
})

comparison.head()
```

##using precision,recall and f1 score

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predicted_jobs))
```

##flask api


```python
from flask import Flask, request, jsonify
import joblib
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

app = Flask(__name__)

# Load the vectorizer (which was saved as 'skill_matching_model.pkl')
vectorizer = joblib.load('skill_matching_model.pkl')

# Load the cleaned dataset to get the job titles and skill text for similarity comparison
df_cleaned = pd.read_csv('cleaned_dataset.csv')

# Ensure 'Skills_Text' column has no NaN values before vectorization
df_cleaned['Skills_Text'] = df_cleaned['Skills_Text'].fillna('')

# Re-create the skill_matrix using the loaded vectorizer and the 'Skills_Text' column
# The vectorizer is already fitted, so we use transform.
skill_matrix = vectorizer.transform(df_cleaned['Skills_Text'])

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    user_input_skills = data.get('skills', [])

    if not user_input_skills:
        return jsonify({'error': 'No skills provided'}), 400

    # Convert the list of skills from the request into a single space-separated string
    # and convert to lowercase for consistency with training data processing.
    user_skills_text = ' '.join(user_input_skills).lower()

    # Transform the user's skills into a TF-IDF vector
    user_vector = vectorizer.transform([user_skills_text])

    # Calculate cosine similarity between user skills and all job skills
    sim = cosine_similarity(user_vector, skill_matrix)

    # Find the index of the job with the highest similarity score
    index = sim.argmax()
    predicted_job = df_cleaned.iloc[index]['Job_Title']
    match_score = sim[0][index] * 100

    return jsonify({'predicted_job': predicted_job, 'match_score': round(match_score, 2)})

if __name__ == '__main__':
    # WARNING: Running Flask app with app.run(debug=True) directly in a Colab cell
    # will block the execution of subsequent cells. To test the API in Colab
    # without blocking, you might consider using tools like ngrok in a separate process
    # or using a framework designed for background serving within notebooks.
    app.run(debug=True, use_reloader=False) # use_reloader=False to prevent issues in Colab
```

```python
# Install Streamlit
!pip install streamlit

import streamlit as st
import joblib
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the vectorizer
vectorizer_streamlit = joblib.load('skill_matching_model.pkl')

# Load the cleaned dataset
df_cleaned_streamlit = pd.read_csv('cleaned_dataset.csv')

# Ensure 'Skills_Text' column has no NaN values before vectorization
df_cleaned_streamlit['Skills_Text'] = df_cleaned_streamlit['Skills_Text'].fillna('')

# Re-create the skill_matrix using the loaded vectorizer
skill_matrix_streamlit = vectorizer_streamlit.transform(df_cleaned_streamlit['Skills_Text'])

st.title("Job Skill Matcher")

user_input_skills = st.text_input("Enter skills separated by comma (e.g., python, sql, ml)")

if st.button("Predict"):
    if user_input_skills:
        # Process user input for skills, converting to lowercase and stripping spaces
        skills_list = [s.strip().lower() for s in user_input_skills.split(',') if s.strip()]
        user_skills_text = ' '.join(skills_list)

        if user_skills_text:
            # Transform the user's skills into a TF-IDF vector
            user_vector = vectorizer_streamlit.transform([user_skills_text])

            # Calculate cosine similarity with the pre-computed skill_matrix
            sim = cosine_similarity(user_vector, skill_matrix_streamlit)

            # Find the index of the best matching job
            index = sim.argmax()
            predicted_job = df_cleaned_streamlit.iloc[index]['Job_Title']
            match_score = sim[0][index] * 100

            st.write(f"**Predicted Job:** {predicted_job}")
            st.write(f"**Match Score:** {round(match_score, 2)}%")
        else:
            st.warning("Please enter valid skills.")
    else:
        st.warning("Please enter some skills.")
```
